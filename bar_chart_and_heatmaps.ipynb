```python
#setup
import pandas as pd 
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
#Dataset from the US Department of Transportation that tracks flight delays - average arrival delay (in minutes) for a different airline and month (all in year 2015)

#filepath
flight_path = "data/flight_delays.csv"

#read file
flight_data = pd.read_csv(flight_path, index_col="Month")
```

```python
flight_data
```

```python
#Bar Chart showing the average arrival delay for F9 by month

#size
plt.figure(figsize=(10,6))

#title
plt.title("Average Arrival Delay for F9 by Month")

#barplot
sns.barplot(x=flight_data.index, y=flight_data['F9'])

#label on vertical axis
plt.ylabel("Arrival delay (in minutes)")
```

```python
#Heatmap

#plot size
plt.figure(figsize=(14,6))

#title
plt.title("Average Arrival Delay for Each Airline, by Month")

#Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)

#label for horizontal axis
plt.xlabel("Airline")
```
